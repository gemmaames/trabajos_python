```python
#1.Importar los datos (web2)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_web2 = pd.read_excel("/content/drive/MyDrive/Python/1-Clases-Python3/8_Octava_Clase/web_2.xlsx")

df_web2.head(5)
```

```python
#2. Sacando el margen y la rentabilidad:
df_web2['Margen'] = df_web2['Ingresos']- df_web2['Gasto']
df_web2['Rentabilidad'] =(df_web2['Margen']/ df_web2['Gasto'])*100

df_web2.head(5)
```

```python
#3.Crear una columna para año y mes (similar al anterior).

df_web2 = df_web2.rename(columns={'dia':'Fecha'})

df_web2.head(5)


```

```python
df_web2['Fecha'] = pd.to_datetime(df_web2['Fecha'])
df_web2['Año'] = df_web2['Fecha'].dt.year
df_web2['Mes'] = df_web2['Fecha'].dt.strftime('%B')

df_web2.head(5)
```

```python
#4.Ingresar en un dataFrame solo los años 2021 y 2022 y visualizar agrupado los margenes y rentabilidad por año

años_seleccionados = [2021, 2022]
df_web2_filtrado = df_web2[df_web2['Año'].isin(años_seleccionados)]
rentabilidad = df_web2_filtrado.groupby('Año')['Rentabilidad'].sum()
margen=  df_web2_filtrado.groupby('Año')['Margen'].sum()

print(rentabilidad)
print(margen)
```
